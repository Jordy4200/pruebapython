## Cargar y visualizar datos

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("computadoras.csv")
print (df.head())
print(df.tail())
```

## Computadoras más baratas por tienda



```python
computadora_mas_barata_por_tienda = df.loc[df.groupby('Tienda')['Precio'].idxmin(), ['Tienda', 'Marca', 'Modelo', 'Precio']]
print("Computadoras más baratas por tienda:")
print(computadora_mas_barata_por_tienda)
```

## Precio promedio por tienda



```python
precio_promedio_por_tienda = df.groupby('Tienda')['Precio'].mean()

print("Precio promedio por tienda:")
print(precio_promedio_por_tienda)
```

## Computadoras más caras por estado



```python
def top_5_computadoras(group):
    return group.nlargest(5, 'Precio')[['Estado', 'Modelo', 'Precio']]

computadoras_mas_caras_por_estado = df.groupby('Estado').apply(top_5_computadoras).reset_index(drop=True)

print("Las 5 computadoras más caras por estado (No se quieren mostrar) ")
print(computadoras_mas_caras_por_estado)
```

## Distribución de almacenamiento



```python
almacenamiento_min = df['Almacenamiento (GB)'].min()
almacenamiento_max = df['Almacenamiento (GB)'].max()
almacenamiento_promedio = df['Almacenamiento (GB)'].mean()

print("Resumen de los valores de almacenamiento (GB):")
print(f"Mínimo: {almacenamiento_min} GB")
print(f"Máximo: {almacenamiento_max} GB")
print(f"Promedio: {almacenamiento_promedio:.2f} GB")
```

## Gráfico de barras: Promedio de precios por marca



```python
promedio = df.groupby('Marca')['Precio'].mean()

plt.figure(figsize=(10, 6))
promedio.plot(kind='bar', color='skyblue')
plt.title('Promedio de Precios por Marca')
plt.xlabel('Marca')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()

```

## Gráfico 2: Tendencia de precios promedio por año

```python
tendencias = df.groupby('Año')['Precio'].mean()

plt.figure(figsize=(10, 6))
tendencias.plot(kind='line', marker='o', linestyle='-', color='blue')
plt.title('Tendencia de Precios Promedio por Año')
plt.xlabel('Año')
plt.ylabel('Precio Promedio')
plt.grid(True)
plt.tight_layout()

plt.show()

```
